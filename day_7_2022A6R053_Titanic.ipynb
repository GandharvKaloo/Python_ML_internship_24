```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
auto=pd.read_csv("titanic.csv")
auto
```

```python
#survived and age
plt.bar(auto["who"], auto["survived"])
plt.show()
```

```python
auto.plot(kind="bar", stacked=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(rc={"axes.facecolor":"#FFFF0F", "axes.grid":True, "xtick.labelsize":10, "ytick.labelsize":3})

sns.barplot(data=auto["survived"], linewidth=0.5, label="Survived")
sns.barplot(data=auto["who"], linewidth=0.05, label="Who")

plt.title("Survived VS Who")
plt.xlabel("Survived", color='g', fontsize=15.5)
plt.ylabel("Who", color='r', fontsize=15.5)
plt.show
```

```python
plt.figure(figsize=(15,6))
sns.set(rc={"axes.facecolor":"#FFFF0F", "axes.grid":True, "xtick.labelsize":30, "ytick.labelsize":30})

sns.lineplot(data=auto["class"], linewidth=1.5, label="CLASS")
sns.lineplot(data=auto["sex"], linewidth=1.5, label="SEX")

plt.title("CLASS VS SEX", fontsize=30)
plt.xlabel("CLASS", color='b', fontsize=15.5)
plt.ylabel("SEX", color='y', fontsize=15.5)
plt.show
```

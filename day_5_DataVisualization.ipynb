```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('student_marks.csv')
sm
```

```python
sm.head(5)
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
sm
```

```python
sm['multiply']=sm[['English','Hindi','Maths','Science','SST']].prod(axis=1)
sm
```

```python
sm['percentage']=sm['Total']/5
sm
```

A=100-80
B=80-60
C=60-50
D=50-40
E=40-30
F=0-30

```python
def clac_grade(p):
    if p<=100 and p>80:
        return 'A'
    if p<=80 and p>60:
        return 'B'
    if p<=60 and p>50:
        return 'C'
    if p<=50 and p>40:
        return 'D'
    if p<=40 and p>30:
        return 'E'
    if p<=30 and p>0:
        return 'F'
    else:
        return 'X'
sm['Grade']=sm['percentage'].apply(clac_grade)
sm    
        
```

```python
plt.figure(figure=(10,6))
for sub in ['English','Hindi','Maths','Science','SST']:
    plt.plot(sm['Name'],sm[sub],marker='o',label=sub)
    plt.title("marker in different subjects")
    plt.xlabel('Name')
    plt.ylabel('Marks')
    plt.legend()
    plt.grid(True)
    plt.show()
```
